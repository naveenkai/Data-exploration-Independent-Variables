```python
mod = ols('Sale_Price ~ Condition_of_the_House',data = data2).fit()
```

```python
Anova_Table = sm.stats.anova_lm(mod,typ=2)
```

```python
Anova_Table
```

```python
mod = ols('Sale_Price ~ Condition_of_the_House',data = data2).fit()
sm.stats.anova_lm(mod,typ=2)
```

```python
mod = ols('Sale_Price ~ Ever_Renovated',data = data2).fit()
sm.stats.anova_lm(mod,typ=2)
```

```python
mod = ols('Sale_Price ~ Zipcode',data = data2).fit()
sm.stats.anova_lm(mod,typ=2)
```

```python
## Binning and creation of dummy variable
```

```python
data2 = pd.get_dummies(data2, columns=['Condition_of_the_House'],drop_first=True)
```

```python
data2.head(10)
```

```python
data2 = pd.get_dummies(data2, columns=['Waterfront_View'],drop_first=True)
```

```python
data2.head()
```

```python
zip_table = data2.groupby('Zipcode').agg({'Sale_Price':'mean'}).sort_values('Sale_Price',ascending=True)
```

```python
zip_table
```

```python
zip_table['zipcode_group'] = pd.cut(zip_table['Sale_Price'],bins=10,
                                    labels=['zipcode_group0',
                                            'zipcode_group1',
                                            'zipcode_group2',
                                            'zipcode_group3',
                                            'zipcode_group4',
                                            'zipcode_group5',
                                            'zipcode_group6',
                                            'zipcode_group7',
                                            'zipcode_group8',
                                            'zipcode_group9'],
                                    include_lowest=True) 
```

```python
zip_table = zip_table.drop(columns='Sale_Price')
```

```python
data2 = pd.merge(data2,
                 zip_table,
                 left_on='Zipcode',
                 how='left',
                 right_index=True)
```

```python
data2 = data2.drop(columns='Zipcode')
```

```python
data2.head(10)
```

```python
data2 = pd.get_dummies(data2,
                        columns= ['zipcode_group'],
                        drop_first=True)
```
